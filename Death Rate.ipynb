# Death Rate in India (2016-2020)

```python
import pandas as pd
import matplotlib.pyplot as mp
%matplotlib inline
```

```python
data = pd.read_csv('Data Sets/Pre Corona Pandemic Datasets/Death Rate.csv')
df = pd.DataFrame(data)
df.head()
```

# Data Preprocessing.

```python
df = df.drop(['$FutureFlag','PM','O3','$TSLCI-PM','$TSUCI-PM','$TSLCI-O3','$TSUCI-O3'],axis=1)
df
```

```python
df.dtypes
```

```python
df['Year'] = pd.to_datetime(df['Year'])
df.dtypes
```

```python
df = df.rename(columns={'$TS-PM':'PM'})
```

```python
df = df.rename(columns={'$TS-O3':'O3'})
```

```python
df
```

# Selecting the data from 2016-2020.

```python
df = df[df['Year']>='2016-01-01']
df
```

```python
df = df.set_index('Year')
df
```

# Perform visualization to generate insight.

```python
y = df['PM']
df['PM'].plot(kind='line',marker='o')
mp.grid(which='both', axis='both',linewidth=0.4)
mp.xlabel('Years',fontsize='large', fontweight='bold')
mp.ylabel('Death Rate',fontsize='large', fontweight='bold')
mp.title('Outdoor particulate matter (Deaths per 100,000)',fontsize='large', fontweight='bold')
for a,b in zip(df.index, y): 
    mp.text(a, b+0.05, str(b),bbox=dict(facecolor='red', alpha=0.8),color='white',fontsize=7,fontweight='bold')
mp.show()  
```

```python
y = df['O3']
df['O3'].plot(kind='line',marker='o')
mp.grid(which='both', axis='both',linewidth=0.4)
mp.xlabel('Years',fontsize='large', fontweight='bold')
mp.ylabel('Death Rate',fontsize='large', fontweight='bold')
mp.title('Outdoor Ozone (Deaths per 100,000)',fontsize='large', fontweight='bold')
for a,b in zip(df.index, y): 
    mp.text(a, b+0.006, str(b),bbox=dict(facecolor='red', alpha=0.8),color='white',fontsize=7,fontweight='bold')
mp.show()
```
