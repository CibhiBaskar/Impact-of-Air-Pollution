# Atmospheric Temperature in NEWS of India (2016-2020)

```python
import pandas as pd
import matplotlib.pyplot as mp
import numpy as np
%matplotlib inline
```

## Chennai

```python
data = pd.read_csv('Data Sets/Temperature Levels/Chennai.csv')
df = pd.DataFrame(data)
df.head()
```

## Delhi

```python
data1 = pd.read_csv('Data Sets/Temperature Levels/Delhi.csv')
df1 = pd.DataFrame(data1)
df1.head()
```

## Mumbai

```python
data2 = pd.read_csv('Data Sets/Temperature Levels/Mumbai.csv')
df2 = pd.DataFrame(data2)
df2.head()
```

## Kolkata

```python
data3 = pd.read_csv('Data Sets/Temperature Levels/Kolkata.csv')
df3 = pd.DataFrame(data3)
df3.head()
```

# Data Preprocessing.

```python
df['City'] = pd.DataFrame(columns={'City'})
df['City'] = df['City'].replace(np.nan,'Chennai')
df
```

```python
df1['City'] = pd.DataFrame(columns={'City'})
df1['City'] = df1['City'].replace(np.nan,'Delhi')
df1
```

```python
df2['City'] = pd.DataFrame(columns={'City'})
df2['City'] = df2['City'].replace(np.nan,'Mumbai')
df2
```

```python
df3['City'] = pd.DataFrame(columns={'City'})
df3['City'] = df3['City'].replace(np.nan,'Kolkata')
df3
```

# Concate all the dataframes to one and manipulate the values.

```python
temp = [df,df1,df2,df3]
df4 = pd.concat(temp)
df4
```

```python
df4 = df4[df4['YEAR']>=2016]
df4
```

```python
df4 = df4.drop('DAY',axis=1)
df4
```

```python
df4 = df4.groupby(['MONTH','YEAR','City'],as_index=False, sort=False).mean()
df4
```

```python
df4.dtypes
```

```python
df4['Date'] = pd.to_datetime(df4[['YEAR', 'MONTH']].assign(Day=1))
df4
```

```python
df4 = df4.drop(['MONTH','YEAR'],axis=1)
df4
```

```python
df4.dtypes
```

```python
df4 = df4.set_index(['Date'])
df4
```

# Perform Visualization to generate insights.

```python
df4 = df4[df4['City']=='Kolkata']
df4.plot(kind='line',marker='.')
mp.grid(which='both', axis='both',linewidth=0.4)
mp.xlabel('Years',fontsize='large', fontweight='bold')
mp.ylabel('Temperature in Fahrenheit',fontsize='large', fontweight='bold')
mp.title('Kolkata Temperature ',fontsize='large', fontweight='bold')
```
