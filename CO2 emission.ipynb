# CO2 Emission in India (2016-2020)

```python
import pandas as pd
import matplotlib.pyplot as mp
%matplotlib notebook
```

```python
data = pd.read_csv("Data Sets/Pre Corona Pandemic Datasets/CO2 Level.csv")
df = pd.DataFrame(data)
df.head()
```

# Data Preprocessing

```python
df.dtypes
```

```python
df['Year'] = pd.to_datetime(df['Year'])
df
```

```python
df.dtypes
```

```python
df = df.set_index('Year')
df
```

# Data Visualization

```python
y = df['CO2 emissions']
df.plot(kind='line',marker='.',color='teal')
mp.grid(which='both', axis='both',linewidth=0.4)
mp.xlabel('Years',fontsize='large', fontweight='bold')
mp.ylabel('Mt of CO2',fontsize='large', fontweight='bold')
mp.title('TOTAL CO2 EMISSIONS IN INDIA',fontsize='large', fontweight='bold')
for a,b in zip(df.index, y): 
    mp.text(a, b+9, str(b),bbox=dict(facecolor='red', alpha=0.7),color='white',fontsize=9,fontweight='bold')
mp.show()    
```
